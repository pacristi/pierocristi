```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()

movies = pd.read_csv("datos.csv").dropna()

language_raw = movies["Language"].tolist()
language = []
for languages in language_raw:
    str_list = languages.split(",")
    language.append(str_list[0])

RT_pctg = movies["Rotten Tomatoes"].tolist()
RT_values = []
for pctg in RT_pctg:
    value = pctg.replace("%","")
    RT_values.append(value)

df_final = pd.DataFrame({"Rotten Tomatoes":RT_values,
            "IMDb":movies["IMDb"].tolist(),
            "Runtime":movies["Runtime"].tolist(),
            "Language":language})

alt.Chart(df_final).mark_circle(size=60).encode(
    x=alt.X('Rotten Tomatoes:Q'),
    y='IMDb',
    size="Runtime",
    color="Language"
)
```
