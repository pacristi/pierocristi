```python
import pandas as pd
import altair as alt

movies = pd.read_csv("datos.csv").dropna()
alt.data_transformers.disable_max_rows()

notas_imdb = movies["IMDb"].tolist()
porcentajes_RT = movies["Rotten Tomatoes"].tolist()

notas_RT = []
for porcentaje in porcentajes_RT:
    nota = int(porcentaje.replace("%",""))/10
    notas_RT.append(nota)

Promedio = []
for nota in notas_imdb:
    Promedio.append((int(nota) + notas_RT[0])/2)
    notas_RT.pop(0)

df_final = pd.DataFrame({"Aos":movies["Year"].tolist(), "Promedio":Promedio})

alt.Chart(df_final).mark_bar().encode(x="Aos", y="Promedio")
    
```
