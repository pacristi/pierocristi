```python
import altair as alt
import pandas as pd

alt.data_transformers.disable_max_rows()
movies = pd.read_csv("datos.csv")
movies =movies.dropna(subset=["Rotten Tomatoes"])
alt.Chart(movies).mark_bar().encode(x=alt.X("Rotten Tomatoes"), y="count()") # no se como ordenar el eje X con los porcentajes
```

```python
lista_netflix = movies.Netflix.tolist()
lista_RT = movies["Rotten Tomatoes"].tolist()
netflix_y_RT = []
for boolean in lista_netflix:
    if boolean == 1:
        netflix_y_RT.append(lista_RT[0])
        lista_RT.pop(0)
    else:
        lista_RT.pop(0)
netflix = pd.DataFrame({"Netflix":netflix_y_RT})
alt.Chart(netflix).mark_bar().encode(x="Netflix", y="count()").configure(countTitle="Nro de Peliculas")
```
