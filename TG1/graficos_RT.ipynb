```python
import altair as alt
import pandas as pd
import copy as cp

alt.data_transformers.disable_max_rows()
movies = pd.read_csv("datos.csv")
movies = movies.dropna(subset=["Rotten Tomatoes"])

RT_pctg = movies["Rotten Tomatoes"].tolist()
RT_values = []
for pctg in RT_pctg:
    value = pctg.replace("%","")
    RT_values.append(value)

df_final = pd.DataFrame({"Rotten Tomatoes":RT_values})

alt.Chart(df_final).mark_bar().encode(x=alt.X('Rotten Tomatoes:Q'), y="count()").configure(countTitle="Nro de Peliculas") # no se como ordenar el eje X con los porcentajes
```

```python
lista_netflix = movies.Netflix.tolist()
lista_RT = cp.copy(RT_values)
netflix_y_RT = []
for boolean in lista_netflix:
    if boolean == 1:
        netflix_y_RT.append(lista_RT[0])
        lista_RT.pop(0)
    else:
        lista_RT.pop(0)
netflix = pd.DataFrame({"Netflix":netflix_y_RT})
alt.Chart(netflix).mark_bar().encode(x=alt.X("Netflix:Q", scale=alt.Scale(domain=(0,100))), y="count()").configure(countTitle="Nro de Peliculas")
```
