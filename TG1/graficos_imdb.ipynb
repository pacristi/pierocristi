```python
import altair as alt
import pandas as pd

alt.data_transformers.disable_max_rows()
movies = pd.read_csv("datos.csv")
lista_imdb = movies.IMDb.tolist()
lista_imdb_redondeada = []
for nota in lista_imdb:
    if str(nota) != "nan":
        lista_imdb_redondeada.append(round(nota))
imdb = pd.DataFrame({"IMDb":lista_imdb_redondeada})
alt.Chart(imdb).mark_bar().encode(x="IMDb", y="count()").configure(countTitle="Nro de Peliculas")
```

```python
lista_netflix = movies.Netflix.tolist()
netflix_y_imdb = []
netflix_y_imdb_redondeado = []
for boolean in lista_netflix:
    if boolean:
        netflix_y_imdb.append(lista_imdb[0])
        lista_imdb.pop(0)
    else:
        lista_imdb.pop(0)
for nota in netflix_y_imdb:
    if str(nota) != "nan":
        netflix_y_imdb_redondeado.append(round(nota))
netflix = pd.DataFrame({"Netflix":netflix_y_imdb_redondeado})
alt.Chart(netflix).mark_bar().encode(x="Netflix", y="count()").configure(countTitle="Nro de Peliculas")

```
